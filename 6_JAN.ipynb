```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Student': ['Suryansh','Ansh','Vivek','Raghav','Arshdeeep','Arjun','Sarthak','Tanish','Aastha','Chirag','Jitender'],
    'Marks': [83, 75, np.nan, 81, 77, 68, 71, 94, 77, 93, 92],
    'Age': [27, 18, 18, 19, 17, np.nan, 18, 19, 16, 19, np.nan],
    'Subject': ['Maths','Chem','Phy','C_lab','Python','Hindi',np.nan,'English','Python','Maths','THF']
}

df = pd.DataFrame(data)

df['Marks'].fillna(df['Marks'].median(), inplace=True)
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Subject'].fillna("AI-ML", inplace=True)


print("Cleaned DataFrame:\n", df)


print("\nShape of DataFrame:", df.shape)
print("\nMissing values count:\n", df.isnull().sum())


plt.figure(figsize=(10,6))
plt.bar(df['Subject'], df['Marks'], color='red')
plt.xlabel("Subject", fontsize=12)
plt.ylabel("Marks", fontsize=12)
plt.title("Marks Distribution by Subject", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
